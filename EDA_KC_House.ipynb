```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
kch = pd.read_csv('kc_house_data.csv')
kch
```

```python
# Melihat kurva dan melihat tipe distribusi data yang condong miring ke kiri
sns.set(color_codes=True)
sns.set_style("whitegrid")
var = ['grade', 'price']
sns.displot(kch[var], kde="reg")
```

```python
# Menghitung Central Tendency menggunakan median(), 
# karena lebih cocok digunakan untuk distribusi miring
kch_central_tendency = kch[['grade','price']].median()
kch_central_tendency
```

```python
# Menggunakan perhitungan standar deviasi
kch_dispersion = kch[['grade','price']].std()
kch_dispersion
```

```python
# Menampilkan outlier menggunakan boxplot
sns.boxplot(x=kch['sqft_living'])
```

```python
# Bangunan yang ada saat ini rata-rata berusia 50 tahun 
sns.boxplot(y=kch['yr_built'])
```
